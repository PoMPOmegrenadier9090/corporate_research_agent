```python
import pandas as pd

# 全角英数字を半角に変換するための変換テーブルを作成
ZEN = "".join(chr(0xff01 + i) for i in range(94))
HAN = "".join(chr(0x21 + i) for i in range(94))
ZEN2HAN = str.maketrans(ZEN, HAN)

# データの読み込み
df = pd.read_csv('agent/tools/Data Report.csv')

# 「-」を空文字に置換
df = df.replace('-', '')

# 「銘柄名」の全角英数字記号を半角に正規化 (カタカナは全角のまま維持)
df['銘柄名'] = df['銘柄名'].astype(str).str.translate(ZEN2HAN)

# 「33業種コード」が空のレコードを削除
# 置換後の空文字、または元のNaNを対象にします
df_filtered = df[df['33業種コード'] != '']
df_filtered = df_filtered.dropna(subset=['33業種コード'])

# 結果の表示
display(df_filtered.head())
print(f"残りの件数: {len(df_filtered)}")
```

```python
df_filtered.to_csv("data.csv", encoding='utf-8')
```
